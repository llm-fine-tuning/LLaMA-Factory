```python
import json
from tqdm import tqdm
import pandas as pd
from transformers import AutoTokenizer
```

```python
tqdm.pandas()
```

```python
tokenizer = AutoTokenizer.from_pretrained('allganize/Llama-3-Alpha-Ko-8B-Instruct')
```

```python
df = pd.read_csv('sample_rag.csv')

df
```

```python
df
```

```python
df['full_prompt'] = df['system_prompt'] + ' ' + df['user_prompt'] + ' ' + df['output']
```

```python
def get_len(text):
  return len(tokenizer.encode(text))
```

```python
df['len'] = df['full_prompt'].progress_apply(get_len)
```

```python
df = df[df['len'] < 8192]
```

```python
df
```

```python
df['instruction'] = df['user_prompt']
```

```python
df['input'] = '' # 라마팩토리 형식을 맞춰주기 위해서 임의로 추가한 열
```

```python
df = df[['instruction', 'input', 'output']]
```

```python
import json

# DataFrame에서 'instruction'과 'output' 열을 딕셔너리 리스트로 변환
data_to_save = df[['instruction', 'input', 'output']].to_dict(orient='records')

# JSON 파일로 저장
with open('rag.json', 'w', encoding='utf-8') as f:
    json.dump(data_to_save, f, ensure_ascii=False, indent=4)
```
